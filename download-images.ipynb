```python
import pandas as pd
import os
import requests
```

```python
df = pd.read_csv("/content/drive/MyDrive/Project3/raw-shoe-data.csv")
df = df.loc[df['gender'] == 'men']
df.head()
```

```python
pd.set_option('display.max_colwidth', None)
data = df[["brand","media.smallImageUrl","shoe"]]
data.columns=["brand","image","shoe"]
data.head()
```

```python
data['shoe'] = [x.replace(" ","-") for x in data['shoe']]
data['shoe'] = [x.replace("/","-") for x in data['shoe']]
data['brand'] = [x.replace("ASICS","Asics") for x in data['brand']]
data.head()
```

```python
data = data.dropna()
data.isna().sum()
```

```python
img_data = data[data['image'].str.contains('https://images')]
```

```python
# short_df = img_data.head(10)
folder = "/content/drive/MyDrive/Project3/images/"
path = []
for i, row in img_data.iterrows():
    filename = folder + str(row.brand)
    
    path.append(f"{filename}/{str(row.shoe)}.jpg")
    
    os.makedirs(filename, exist_ok=True)
    
    url = row.image
    # print(f"Downloading {url} to {filename}")
    r = requests.get(url, allow_redirects=True)
    with open(f"{filename}/{str(row.shoe)}.jpg", 'wb') as f:
        f.write(r.content)
```
