```python
import pandas as pd
import requests
import json
```

```python
url = 'https://api.thesneakerdatabase.com/v1/sneakers?limit=100&page='
data = []
for x in range(100):
    response = requests.get(url+str(x)).json()
    for result in response['results']:
        data.append(result)
```

```python
raw = pd.json_normalize(data)
raw.shape
```

```python
raw.to_csv('raw-shoe-data.csv', index=False)
```

```python
df = pd.read_csv('raw-shoe-data.csv')
```

```python
df[['primary','secondary','tertiary']] = df['colorway'].str.split("/", n=2, expand=True)
df.head()
```

```python
df.isna().sum()
```

```python
df = df[df.retailPrice != 0]
df.head()
```

```python
df['retailPrice'].value_counts().sort_index()
```

```python
bins = [0,100,150,200,250,300,1010]
labels = ['<100','100-150','150-200','200-250','250-300','>300']
df['priceGroup'] = pd.cut(df.retailPrice, 
                          bins=bins, 
                          labels=labels)
df.head()
```

```python
df['priceGroup'].value_counts().sort_index()
```
