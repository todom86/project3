```python
import pandas as pd
import os
import requests
```

```python
df = pd.read_csv("raw-shoe-data.csv")
df = df.loc[df['gender'] == 'men']
df
```

```python
pd.set_option('display.max_colwidth', None)
data = df[["brand","media.smallImageUrl","shoe"]]
data.columns=["brand","image","shoe"]
data.head()
```

```python
data['shoe'] = [x.replace(" ","-") for x in data['shoe']]
data
```

```python
data = data.dropna()
```

```python
img_data = data[data['image'].str.contains('https://images')]
```

```python
path = []
def download(row):
    filename = os.path.join('images', str(row.brand),
                           str(row.shoe) + '.jpg')
    
    path.append(filename)
    
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    
    url = row.image
#     print(f"Downloading {url} to {filename}")
    r = requests.get(url, allow_redirects=True)
    with open(filename, 'wb') as f:
        f.write(r.content)
```

```python
img_data
```

```python
img_data.apply(download, axis=1)
```
